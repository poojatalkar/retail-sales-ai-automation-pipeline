```python
import pandas as pd

# Open raw file
df = pd.read_csv('retail_sales_messy_100K.csv')
print(f"Before: {len(df)} rows")
```

Fix 1 — Remove duplicate orders

```python
df = df.drop_duplicates(subset=['OrderID'])
```

Fix 2 — Remove negative quantity rows

```python
df = df[df['Quantity'] > 0]
```

Fix 3 — Remove zero price rows

```python
df = df[df['UnitPrice'] > 0]
```

Fix 4 — Fix discount scale if values are like 50, 30

```python

if df['Discount'].max() > 1:
    df['Discount'] = df['Discount'] / 100
```

Fix 5 — Fill empty CustomerID with UNKNOWN

```python
df['CustomerID'] = df['CustomerID'].fillna('UNKNOWN')
```

Fix 6 — Delete rows where ProductName is empty

```python
df = df.dropna(subset=['ProductName'])
```

 Fix 7 — Remove extra spaces from ProductName

```python
df['ProductName'] = df['ProductName'].str.strip()
```

Fix 8 — Fix region name capitalization

```python
df['Region'] = df['Region'].str.strip().str.title()
```

Save the clean file

```python
df.to_csv('retail_sales_clean.csv', index=False)
print("Clean file saved!")
```
