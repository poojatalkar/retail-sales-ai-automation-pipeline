## find the problems: Exploratory Data Analysis

```python
# How many empty cells in each column?
print(df.isnull().sum())
```

```python
# How many duplicate orders?
print(df.duplicated(subset=['OrderID']).sum())
```

```python
# Any negative quantities?
print(df[df['Quantity'] < 0].shape[0])
```

```python

# Any zero prices?
print(df[df['UnitPrice'] <= 0].shape[0])
```

```python

# What does discount look like?
print(df['Discount'].min())
print(df['Discount'].max())
```
