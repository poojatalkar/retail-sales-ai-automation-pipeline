```python
from groq import Groq
import pandas as pd

df = pd.read_csv('C:/retail_sales_clean.csv')

total_revenue = round(df['Revenue'].sum(), 2)
total_profit  = round(df['Profit'].sum(), 2)
margin        = round((total_profit/total_revenue)*100, 2)
total_orders  = df['OrderID'].nunique()
aov           = round(total_revenue/total_orders, 2)
top_category  = df.groupby('Category')['Revenue'].sum().idxmax()
online_pct    = round(
    df[df['Channel']=='Online']['Revenue'].sum()
    /total_revenue*100, 1)

print("--- Your KPIs ---")
print("Total Revenue =", total_revenue)
print("Profit Margin =", margin, "%")
print("Total Orders  =", total_orders)
print("AOV           =", aov)
print("Top Category  =", top_category)
print("Online %      =", online_pct, "%")
print("")

prompt = f"""
You are a retail business analyst.
Give me 3 simple insights and 2 recommendations
based on these numbers:

Total Revenue   = ${total_revenue:,}
Profit Margin   = {margin}%
Total Orders    = {total_orders:,}
Avg Order Value = ${aov}
Top Category    = {top_category}
Online Revenue  = {online_pct}%

Keep it short and simple.
"""

client = Groq(api_key="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")

response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[{"role": "user", "content": prompt}]
)

print("--- AI Generated Insights ---")
print(response.choices[0].message.content)
```
